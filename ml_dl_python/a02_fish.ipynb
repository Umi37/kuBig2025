```python
import matplotlib.pyplot as plt
plt.scatter(bream_length, bream_weight)
plt.xlabel('Length (cm)')
plt.ylabel('Weight (g)')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight, color='orange')
plt.xlabel('Length (cm)')
plt.ylabel('Weight (g)')
plt.show()
```

```python
length = bream_length + smelt_length
weight = bream_weight + smelt_weight
fish_data = [[l, w] for l, w in zip(length, weight)]
fish_labels = [1] * len(bream_length) + [0] * len(smelt_length)
for data, label in zip(fish_data, fish_labels):
    print(f"DATA : {data}, LABEL : {label}")
```

```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

fish_data = np.array(fish_data)
kn = KNeighborsClassifier()
kn.fit(fish_data, fish_labels)
```

```python
kn.score(fish_data, fish_labels)
# 위 코드의 실행결과 = 1.0 (모든 훈련 데이터에 대해 정확하게 분류했다는 의미)
```

```python
kn.predict(np.array([[20.0, 300.0], [13.0, 20.0]]))
# 위 코드의 실행결과 = [1, 0] (첫 번째는 도미, 두 번째는 빙어로 분류됨)
```
